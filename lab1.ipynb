```python
### comments reiterate what I've written in the word document
### Below is the algortihm with no optimizations for comparison

%matplotlib inline
import matplotlib.pyplot as plt

checked=[]
total_steps=0
steps=[]

for i in range (1,1001):
    n=i
    step_count=0

    while(n!=1):
        
        if (n%2==0):
            n = n/2
            
        else:
            n = 3*n+1

        step_count += 1

    checked.append(i)
    total_steps += step_count
    steps.append(total_steps)
    
print("total iterations:",total_steps)
plt.scatter(checked,steps)
plt.show()
```

```python
###Question 1
### Algortithm with optimization to reduce redundancy moving from 1 to 1000
### If a number in a sequence has been shown not to create an infinite loop
### when it is the starting value, the sequence terminates
### This is not the best way to do this but in the question it says to only 
### save the starting values

%matplotlib inline
import matplotlib.pyplot as plt

value_range=[]
total_steps=0
steps=[]

for i in range (1,1001):
    all_good=0
    n=i
    step_count=0

    while(n!=1):

        for k in range(i-1):
            if (n==checked[k]):
                all_good = 1
                break

        if (all_good==1):
            break

        if (n%2==0):
            n = n/2
        else:
            n = 3*n+1

        step_count += 1

    value_range.append(i)
    total_steps += step_count
    steps.append(total_steps)

print("total iterations:",total_steps)
plt.scatter(value_range,steps)
plt.show()
```

```python
### Question 2
### Same as above however, starting with the value 1000 and moving down to 1
### From counting total steps it has become clear that moving up is 
### (the way it's done in Question 1)
### is much better than counting down from 1000
### This is because the higher numbers take additional steps to solve 
### Essentially we're doing the work for the more computationally
### expensive numbers in order to reduce work for the simpler ones
### Please note that if I were to save all of the numbers rather than
### just the starting values, the order in which the list is traversed
## does NOT matter

%matplotlib inline
import matplotlib.pyplot as plt

value_range=[]
checked=[-1]*1001
total_steps=0
steps=[]
n=0

for i in range (1000,1,-1):
    all_good=0
    n=i  
    step_count=0

    while(n!=1):
          
        if(n<1001):
            if (checked[int(n)]!=-1):
                break
                
        if (n%2==0):
            n = n/2
        else:
            n = 3*n+1

        step_count += 1

    checked[i]=i
    value_range.append(i)
    total_steps += step_count
    steps.append(total_steps)

print("total iterations:",total_steps)
plt.scatter(value_range[::-1],steps)
plt.show()
```

```python
### Question 3
### In the next few blocks of code I will try different upper bounds
### Algortithm with additional optimization to reduce redundancy
### Any number that is a part of a sequence that completes is logged
### and if it is hit during a later sequence the sequence terminates

%matplotlib inline
import matplotlib.pyplot as plt

checked=[-1]*1000
total_steps=0
steps=[]
value_range=[]

for i in range (1,1001):
    all_good=0
    n=i
    step_count=0
    log=[]

    while(n!=1):
        
        if(n<1000):           
            if(checked[int(n)]==n):
                break

        if (n%2==0):
            n = n/2
        else:
            n = 3*n+1
        log.append(n)
        step_count += 1
        
    for k in range (len(log)):
        if(log[k]<1000):
            checked[int(log[k])]=log[k]
        
    value_range.append(i)
    total_steps += step_count
    steps.append(total_steps)
    
print("total iterations:",total_steps)

plt.scatter(value_range,steps)
plt.show()
```
